## Google Tag Manager & Analytics Expertise: IngredientOnline 

At **IngredientOnline**, I was deeply involved in ensuring that our digital footprint resonated with our business objectives and data requirements. Below is a detailed breakdown of my work with Google Tag and Google Analytics at the company.

### 1. Business Scenario of IngredientOnline:

IngredientOnline is a B2B platform connecting raw material suppliers with food manufacturers. To improve our understanding of our users' behavior and to enhance our service delivery, we sought to track multiple touchpoints of our user journey via Google Tag Manager and Google Analytics.

### 2. Identifying Key Performance Indicators (KPIs):

After extensive discussions with both internal teams (Data Scientist Team) and external teams (Seller, Marketing, Buyer Teams, and Technical Teams), we identified a list of KPIs and data collection needs. Some of these included:
- Begin checkout
- Cart icon click
- Newsletter subscription
- Order placement
- Seller sign-up
- And more...

### 3. Google Tag Event Creation:

Based on the requirements, I created specific Google event tags, leveraging Google triggers and Google variables. These variables were custom-defined using advanced JavaScript code to capture precise user actions and data. Visual representations of these tags can be seen below:






### 4. Testing and Verification:

All the configured Google tags were rigorously tested in the Google Tag Manager's preview (debug) mode to ensure accuracy and desired output. This allowed for real-time testing and modifications, ensuring that data capture was in line with our objectives.



### 5. Data Integration and Live Application:

Once satisfied, the Google tag events were set live in GTM. Within 48 hours, the data from these tags started reflecting in Google Analytics. This data, once captured in Analytics, was ready to be pulled and pushed into our AWS database. This rich dataset was then made available for exploratory data analysis and fed into advanced machine learning models to derive actionable insights for IngredientOnline.
